# Reference Frame Calibration

Source Extractor provides a catalog file (`.ctlg`) with the estimated pixel position, RA/DEC position, amplitude, and full width half max.

This algorithm pushes the source extractor catalog into a dataframe and uses those positions to subtract a Gaussian profile (currently) from each source. 
    - Attempted to use an Airy function, but the residuals were a factor of 3-5 worse.

Stage 1: Sanity Check
    1. Postage stamp each star in the reference frame source extractor catalog
    2. Fit a gaussian to the postage stamps
    3. Store those center, width amplitude into a new catalog dataframe
    4. Build a reference template to subtract from the reference frame (sanity chack)

Stage 2: Results
    1. Postage stamp each star location in the science frame based on positions in the reference frame
    2. Fit a Gaussian (later an Airy Disk) to each science postage stamp
    3. Store the centers, widths, amplitudes into a new catalog dataframe
    4. Buidl a reference template to subtract from the science frame (result)

````python
def read_fits_as_dataframe(filename, index_columns=None):
    # This is the way to read the FITS data into a numpy structured array
    # (using astropy.io.fits.getdata didn't work out of the box
    # because it gives a FITSRec)
    # https://github.com/cdeil/test_fits_to_pandas
    table = Table.read(filename)
    # Fix byte order.
    # See https://github.com/astropy/astropy/issues/1156
    data  = {}
    for key in table.keys():
        if np.ndim(table.as_array()[key]) == 2:
            for k in range(table.as_array()[key].shape[1]):
                data[key + '_' + str(k)] = table.as_array()[key][:,k].byteswap().newbyteorder()
        else:
            data[key] = table.as_array()[key].byteswap().newbyteorder()
    
    df    = pd.DataFrame.from_records(data)
    if index_columns is not None:
        # Strip whitespace for string columns that will become indices
        for index_column in index_columns:
            df[index_column]  = df[index_column].map(str.strip)
        df  = df.set_index(index_columns)
    return df
````

````python
def postage_stamp_centering(image, ctlg, nFWHM=2, yinds=None, xinds=None, subframe=None, returnStamps=False):
    # returnStamps=False
    # nFWHM =2  
    # yinds = None
    # xinds = None
    # subframe  = [[1100,1200], [1250,1350]]
    # image = refimage.data
    # ctlg  = refctlg
    y,x     = 0,1
    fitter  = fitting.LevMarLSQFitter()
    # if subframe is None:
    #     subframe = [[0,image.shape[0]], [0,image.shape[1]]]

    if yinds is None or xinds is None:
        yinds, xinds  = np.indices(image.shape)

    if returnStamps:
        postageStamps = []
    
    ctlgOut = pd.DataFrame(columns = ['airy_amplitude', 'x_center', 'y_center', 'airy_width', 'background_amplitude',
                                      'Y_IMAGE','X_IMAGE','FLUX_MAX','FWHM_IMAGE','BACKGROUND', 'chisq_old', 'chisq_new'])
    
    start0  = time()
    for k in range(len(ctlg['Y_IMAGE'])):
        if ctlg['Y_IMAGE'][k] > subframe[y][0] and \
           ctlg['Y_IMAGE'][k] < subframe[y][1] and \
           ctlg['X_IMAGE'][k] > subframe[x][0] and \
           ctlg['X_IMAGE'][k] < subframe[x][1]:
                start1  = time()
                # Allocate input location of postage stamp and initial Airy inputs
                # FITS files and PYTHON encode the first index differently
                #  resulting in a difference of -1

                Amplitude_Now   = ctlg['FLUX_MAX'][k]
                FWHM_Now        = ctlg['FWHM_IMAGE'][k]*1.22
                ycenter_Now     = ctlg['Y_IMAGE'][k] - 1
                xcenter_Now     = ctlg['X_IMAGE'][k] - 1
                background_Now  = ctlg['BACKGROUND'][k]
                
                # HACKS: For some reason, some of the widths are negative
                FWHM_Now        = abs(FWHM_Now)
                
                ylower, yupper  = ycenter_Now - nFWHM*FWHM_Now, ycenter_Now + nFWHM*FWHM_Now
                xlower, xupper  = xcenter_Now - nFWHM*FWHM_Now, xcenter_Now + nFWHM*FWHM_Now
                
#                 print k, Amplitude_Now, xcenter_Now, ycenter_Now, FWHM_Now, time()-start1, time()-start0
                
                postageStampNow = image[ylower:yupper, xlower:xupper]

                if returnStamps:
                    postageStamps.append(postageStampNow)
                
                # print 'Organize init Params for Models'
                model0  = models.AiryDisk2D(Amplitude_Now, \
                                            xcenter_Now  , \
                                            ycenter_Now  , \
                                            FWHM_Now    )+ \
                          models.Const2D(background_Now)

                model1  = fitter(model0, 
                                 xinds[ylower:yupper, xlower:xupper], 
                                 yinds[ylower:yupper, xlower:xupper], 
                                 postageStampNow)
                
                outputDict                          = {}
                outputDict['airy_amplitude']        = model1.amplitude_0[0]
                outputDict['x_center']              = model1.x_0_0[0]
                outputDict['y_center']              = model1.y_0_0[0]
                outputDict['airy_width']            = model1.radius_0[0]
                outputDict['background_amplitude']  = model1.amplitude_1[0]
                outputDict['Y_IMAGE']               = ycenter_Now# + ylower
                outputDict['X_IMAGE']               = xcenter_Now# + xlower
                outputDict['FLUX_MAX']              = Amplitude_Now
                outputDict['FWHM_IMAGE']            = FWHM_Now
                outputDict['BACKGROUND']            = background_Now
                outputDict['chisq_old']             = ((postageStampNow - model0(xinds[ylower:yupper, xlower:xupper],
                                                                                 yinds[ylower:yupper, xlower:xupper]))**2.).sum()
                outputDict['chisq_new']             = ((postageStampNow - model1(xinds[ylower:yupper, xlower:xupper],
                                                                                 yinds[ylower:yupper, xlower:xupper]))**2.).sum()
                
                ctlgOut.loc[len(ctlgOut)] = outputDict
                
    if returnStamps:
        return ctlgOut, postageStamps, model1, model0
    else:
        return ctlgOut
````
